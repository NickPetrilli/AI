```python
# Part III - Read in the XOR data
import pandas as pd

df = pd.read_excel(filename)

print(df)
```

```python
import numpy as np

xInputs = []
yLabels = []

for index in range(len(df)):
  row = df.iloc[index]
  # Access row data using row['column_name']
  xInputs.append([row['X1'], row['X2'], row['X3']])
  yLabels.append(row['Y1'])

# print(xInputs)
# print(yLabels)

X = np.array(xInputs)
y = np.array(yLabels)

print(X)
print(y)
```

```python
# Import the modules from Keras which we can get from Keras or Keras of tensorflow.
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.initializers import Constant
```

```python
# Part IV - Build the model.

# Create the model with three inputs and one output
model = Sequential()

# Default bias is 0.0 -- we set it to 1.0 as an experiment.
model.add(Dense(3, input_dim=3, activation='relu', bias_initializer=Constant(1.0)))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile the model parameters
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# This class definition may give "keras not defined warning..." To fix, import Keras
# from tensorflow as shown above.
class DiagnosticCallback(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        if (epoch + 1) % 100 == 0:
            print(f"Epoch {epoch + 1}: loss = {logs['loss']:.4f}, accuracy = {logs['accuracy']:.4f}")
```

```python
# Part V - Train the model

# Set verbose=0 to suppress ALL diagnostics or use the diagnostic callback defined
# above -- apparently there's no other way to ask for SOME periodic diagnostics.
NUM_EPOCHS = 2000
model.fit(X, y, epochs=NUM_EPOCHS, verbose=0,callbacks=[DiagnosticCallback()])
```

```python
# Make predictions against all the actual inputs.
# Sometimes you might a warning, "WARNING:tensorflow:5 out of the last 5 calls...""
predictions = model.predict(X)
```
